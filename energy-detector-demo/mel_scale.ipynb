```python
import numpy as np
import matplotlib.pyplot as plt

# 绘制梅尔刻度
points = 2000
hz_x = np.linspace(0,8000,points)
# mel_y = 2595 * np.log10(1 + hz_x / 700)
mel_y = 1127 * np.log(1 + hz_x / 700)

fig, ax = plt.subplots()
ax.plot(hz_x, mel_y, c='r')
ax.set_xlabel('Hz')
ax.set_ylabel('Mel')
ax.grid()
plt.title("Mel Scale")
plt.show()
```
